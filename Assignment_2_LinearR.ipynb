Use data from August 1st 2021 up to September 30th 2021 as your training set. Your test set will contain the data from October 1st 2021 to October 8th 2021 (8 days in total).

Use your model to predict the number of vaccinations in South Korea for the ten days in the test set (not included in your training set).


```python
df1.head()
```

```python
df3= df1.loc[df1["location"] == "South Korea", :]
```

```python
df3
```

```python
kor_vacc = df3.groupby('date')['total_vaccinations'].sum()
```

```python
kor_vacc
```

```python
kor_vac = kor_vacc.to_frame().reset_index(drop=False)
kor_vac.head()
```

```python
#df = pd.DataFrame(data)
start_date = '2021-08-01'
end_date = '2021-09-30'
mask = (kor_vac['date'] > start_date) & (kor_vac['date'] <= end_date)
  
train_data = kor_vac.loc[mask]
display(train_data)
```

```python
train_data['date'] = pd.to_datetime(train_data['date'])
today = pd.Timestamp(datetime.today().date())
train_data['Days_Since'] = (today - train_data['date']).dt.days
train_data.head()
```

```python
start_date = '2021-10-01'
end_date = '2021-10-08'
mask = (kor_vac['date'] > start_date) & (kor_vac['date'] <= end_date)
  
test_data = kor_vac.loc[mask]
display(test_data)
```

```python
test_data['date'] = pd.to_datetime(test_data['date'])
today = pd.Timestamp(datetime.today().date())
test_data['Days_Since'] = (today - test_data['date']).dt.days
```

```python
test_data
```

```python
X_train = train_data[['Days_Since']]
#df['date'] = df['date'].astype(float)
y_train = train_data[['total_vaccinations']]
print(X_train.head())
```

```python
X_test = test_data[['Days_Since']]
#df['date'] = df['date'].astype(float)
y_test = test_data[['total_vaccinations']]
#elastic net
print(X_test.head())
```

```python
from sklearn.linear_model import ElasticNet
from sklearn.metrics import mean_squared_error

# Create an Elastic Net regression model
en = ElasticNet(alpha=0.1, l1_ratio=0.5)

# Train the model on the training set
en.fit(X_train, y_train)

# Make predictions on the test set
y_pred = en.predict(X_test)
print("Predicted Value for X_test:", y_pred)
# Evaluate the model using mean squared error (MSE)
mse = mean_squared_error(y_test, y_pred)
print("Mean squared error:", mse)
```
