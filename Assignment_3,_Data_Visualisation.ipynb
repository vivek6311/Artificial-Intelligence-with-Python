```python
import pandas as pd
import numpy as np
#import matplotlib as plt
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
google_data= pd.read_csv('/content/google.txt', sep='\t')
temp_data= pd.read_csv('/content/ny.txt', sep='\t')
yahoo_data= pd.read_csv('/content/yahoo.txt', sep='\t')
```

```python
# Create a figure and axes object
fig, ax = plt.subplots(figsize=(15, 10))

# Plot the Yahoo stock prices
ax.plot(yahoo_data["Modified Julian Date"], yahoo_data["Stock Value"],color="purple", label="Yahoo")

# Plot the Google stock prices
ax.plot(google_data["Modified Julian Date"], google_data["Stock Value"], color="green",label="Google")

# Set the x-axis label and tick labels
ax.set_xlabel("Date")
ax.set_xticklabels(yahoo_data["Modified Julian Date"], rotation=90)

# Set the y-axis label and tick labels
ax.set_ylabel("Stock Price")
ax.set_yticklabels(["${:.0f}".format(price) for price in ax.get_yticks()])

# Create a second y-axis for the temperature data
temp_ax = ax.twinx()
temp_ax.plot(temp_data["Modified Julian Date"], temp_data["Max Temperature"], color="blue", label="Temperature")
temp_ax.set_ylabel("Temperature (F)")

# Add a legend
ax.legend(loc="upper left")
temp_ax.legend(loc="upper right")

# Set the title
plt.title("Yahoo and Google Stock Prices with New York Temperature")

# Display the plot
plt.show()
```
