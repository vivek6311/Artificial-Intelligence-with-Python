Determine the theoretical time complexity and verify the empirical time complexity for the following algorithms. 

```python
import time
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
```

### Algorithm 1 :

Find the theoretical time complexity of the below code for this algorithm.

Mention how you have arrived at this Big-O complexity in the report. 

```python
def algo1(lis_arr):
    n = len(lis_arr)
    for i in range(n):
        for j in range(n - i - 1):
            if lis_arr[j] > lis_arr[j + 1]:
                lis_arr[j+1], lis_arr[j] = lis_arr[j], lis_arr[j+1]
                
    return lis_arr
```

```python
# Run this code block to execute this algorithm for various input list sizes

algo1_time = []

datapoints1 = np.arange(1,2000,100)

for i in range(len(datapoints1)):
    x = random.randint(1,100,datapoints1[i])
    start = time.time()
    algo1(x)
    algo1_time.append(time.time() - start)
```

```python
# Run this block and verify the time complexities you have calculated theoretically with the plots generated

plt.figure(figsize=(12,6))
plt.plot(datapoints1, algo1_time, label='Algorithm 1 time complexity')
plt.ylabel('Time')
plt.xlabel('Number of elements in the array')
plt.legend()
plt.show()
```

### Algorithm 2

Find the theoretical time complexity of the below code for this algorithm.

Mention how you have arrived at this Big-O complexity in the report. 

```python
def algo2(n):
    if((n == 1) or (n == 2)):
        return  1
    if(n > 2):
        return algo2(n - 1) + algo2(n - 2)
    
```

### Algorithm 3 :

Find the theoretical time complexity of the below code for this algorithm.

Mention how you have arrived at this Big-O complexity in the report. 

```python
def algo3(lis_arr, target, low = None, high = None):
    if low is None:
        low = 0
    if high is None:
        high = len(lis_arr) - 1

    if high < low:
        return -1
    midpoint = (low + high) // 2

    if lis_arr[midpoint] == target:
        return midpoint
    elif target < lis_arr[midpoint]:
        return algo3(lis_arr, target, low, midpoint-1)
    else:
        return algo3(lis_arr, target, midpoint+1, high)
```

```python
# Run this code block to execute this algorithm for various input list sizes

algo3_time = []

datapoints3 = np.arange(1,100000,1000)

for i in range(len(datapoints3)):
    x = random.randint(1,100,datapoints3[i])
    start = time.time()
    algo3(x, 456)
    algo3_time.append(time.time() - start)
```

```python
# Run this block and verify the time complexities you have calculated theoretically with the plots generated

plt.figure(figsize=(12,6))
plt.plot(datapoints3, algo3_time, label='Algorithm 3 time complexity')
plt.ylabel('Time')
plt.xlabel('Number of elements in the array')
plt.legend()
plt.show()
```

### Algorithm 4 :

```python
def algo4(lis_arr, target):
    for i in range(len(lis_arr)):
        if lis_arr[i] == target:
            return i
    return -1
```

### Algorithm 5 : 

Find the theoretical time complexity of the below code for this algorithm.

Mention how you have arrived at this Big-O complexity in the report. 

```python
def algo5_helper(left, right):
    output = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            output.append(left[i])
            i += 1
        else:
            output.append(right[j])
            j += 1
    output.extend(left[i:])
    output.extend(right[j:])
    return output


def algo5(lis_arr):
    list_length = len(lis_arr)
    if list_length == 1:
        return lis_arr
    else:
        mid_point = list_length // 2
        left_partition = algo5(lis_arr[:mid_point])
        right_partition = algo5(lis_arr[mid_point:])
        return algo5_helper(left_partition, right_partition)
        
```

```python
# Run this code block to execute this algorithm for various input list sizes

algo5_time = []

datapoints5 = np.arange(1,10000,500)

for i in range(len(datapoints5)):
    x = random.randint(1,100,datapoints5[i])
    start = time.time()
    algo5(x)
    algo5_time.append(time.time() - start)
```

```python
# Run this block and verify the time complexities you have calculated theoretically with the plots generated

plt.figure(figsize=(12,6))
plt.plot(datapoints5, algo5_time, label='ALgorithm 5 time')
plt.ylabel('Time')
plt.xlabel('Number of elements in the array')
plt.legend()
plt.show()
```
